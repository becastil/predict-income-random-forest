```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
# Step 1 ,2, 4

# Load the dataset, fixing extra spaces after commas
# income_data = pd.read_csv(r"C:\Users\becas\OneDrive\Documents\predict-income-random-forest\income.csv",header=0, delimiter = ", ")
income_data = pd.read_csv(r"C:\Users\becas\OneDrive\Documents\predict-income-random-forest\income.csv",header=0)

# Preview the first few rows
income_data.head()
```

```python
# Step 3
print(income_data.iloc[0])
```

```python
# Step 5
labels = income_data[["income"]]
labels.head()
```

```python
# Step 6, 10

# Step 12
income_data["sex-int"] = income_data["sex"].apply(lambda row: 0 if row == "Male" else 1)
# Step 15
income_data["country-int"] = income_data["native-country"].apply(lambda row: 0 if row == "United-States" else 1)

# Select relevant features (columns) to use as input data
data = income_data[["age", "capital-gain", "capital-loss", "hours-per-week", "sex-int", "country-int"]]

# Preview the selected features
data.head()
```

```python
# Step 7: Split features and labels into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, random_state = 1) 
```

```python
# Step 8
forest = RandomForestClassifier(random_state = 1)
```

```python
# Step 9: Train the Random Forest using the training data and labels
forest.fit(train_data, train_labels)
```

```python
# Step 11 
accuracy = forest.score(test_data, test_labels)
print(f"Model accuracy with 'country-int': {accuracy:.2%}")
```
